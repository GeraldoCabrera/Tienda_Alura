### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Clasificamos las tiendas
valores = ['tienda_1', 'tienda2', 'tienda3', 'tienda4']
for df, valor in zip([tienda, tienda2, tienda3, tienda4], valores):
    df['tienda'] = valor
```

```python
tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])
tiendas.columns = tiendas.columns.str.lower()
tiendas = tiendas.rename(columns={'fecha de compra':'fecha',
                                  'categoría del producto':'categoria_producto',
                                  'método de pago':'metodo_pago',
                                  'costo de envío':'costo_envio',
                                  'lugar de compra':'lugar_compra',
                                  'cantidad de cuotas':'cantidad_cuotas'})

tiendas['fecha'] = pd.to_datetime(tiendas['fecha'], format='%d/%m/%Y')
tiendas.head()
```

# 2. Ventas por categoría

```python
print("Ventas por categoria de producto:")
display(tiendas['categoria_producto'].value_counts())
display(f"Cantidad de categorias: {tiendas['categoria_producto'].nunique()}")
```

```python
# Mayor cantidad de categorias vendida por tienda

valores = ['tienda1','tienda2','tienda3','tienda4']

print("Categorias mas vendidas por tienda:")
for tienda in valores:
    top = tiendas[tiendas['tienda'] == tienda]['categoria_producto'].value_counts().head(1)
    categoria = top.index[0]
    cantidad = top.values[0]
    display(f"{tienda} - {categoria}: {cantidad}")

print("Categorias menos vendidas por tienda: ")
for tienda in valores:
    bottom = tiendas[tiendas['tienda'] == tienda]['categoria_producto'].value_counts().tail(1)
    categoria = bottom.index[0]
    cantidad = bottom.values[0]
    display(f"{tienda} - {categoria}: {cantidad}")
```

# 4. Productos más y menos vendidos

```python
valores = ['tienda1','tienda2','tienda3','tienda4']

print("Productos mas vendidos por tienda:")
for tienda in valores:
    top = tiendas[tiendas['tienda'] == tienda]['producto'].value_counts().head(1)
    categoria = top.index[0]
    cantidad = top.values[0]
    display(f"{tienda} - {categoria}: {cantidad}")

print("Productos menos vendidos por tienda: ")
for tienda in valores:
    bottom = tiendas[tiendas['tienda'] == tienda]['producto'].value_counts().tail(1)
    categoria = bottom.index[0]
    cantidad = bottom.values[0]
    display(f"{tienda} - {categoria}: {cantidad}")
```
